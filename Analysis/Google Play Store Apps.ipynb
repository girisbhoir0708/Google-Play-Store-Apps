```python
def convert_into_bytes(column_name):
    if isinstance(column_name, str):
        if 'k' in column_name:
            return float(column_name.replace("k", "")) * 1024
        elif 'M' in column_name:
            return float(column_name.replace("M", "")) * 1024 * 1024
        elif 'Varies with device' in column_name:
            return np.nan
    return column_name
```

```python
df['Size'] = df['Size'].apply(convert_into_bytes)
```

```python
category_counts = df_ver3['Category'].value_counts()
```

```python
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar', color='skyblue')
plt.title('Count of Apps by Category')
plt.xlabel('Category')
plt.ylabel('Number of Apps')
plt.xticks(rotation=90)
plt.tight_layout()
```

```python
df_ver3['Installs'] = df_ver3['Installs'].replace('[\+,]', '', regex=True).astype(int)
```

```python
category_installs = df_ver3.groupby('Category')['Installs'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12, 8))
category_installs.head(10).plot(kind='bar', color='lightgreen')
plt.title('Top Categories by Total Downloads')
plt.xlabel('Category')
plt.ylabel('Total Installs')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df_ver3['Reviews'] = pd.to_numeric(df_ver3['Reviews'], errors='coerce')
df_ver3['Rating'] = pd.to_numeric(df_ver3['Rating'], errors='coerce')
```

```python
df_ver3_clean = df_ver3.dropna(subset=['Reviews', 'Rating'])
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_ver3_clean['Reviews'], df_ver3_clean['Rating'], alpha=0.5, color='orange')
plt.title('Ratings vs. Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Rating')
plt.tight_layout()
plt.show()
```

```python
df_ver3['Price'] = df_ver3['Price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df_ver3['Installs'] = df_ver3['Installs'].replace('[\+,]', '', regex=True).astype(int)
```

```python
df_ver3_clean = df_ver3.dropna(subset=['Price', 'Installs'])
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_ver3_clean['Price'], df_ver3_clean['Installs'], alpha=0.5, color='purple')
plt.title('Price vs. Number of Installs')
plt.xlabel('Price (USD)')
plt.ylabel('Number of Installs')
plt.tight_layout()
plt.show()
```

```python
average_rating_by_category = df_ver3.groupby('Category')['Rating'].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12, 8))
average_rating_by_category.plot(kind='bar', color='skyblue')
plt.title('Average Rating by Category')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Split the dataset into free and paid apps
free_apps = df_ver3[df_ver3['Type'] == 'Free']
paid_apps = df_ver3[df_ver3['Type'] == 'Paid']
```

```python
average_values = pd.DataFrame({
    'Rating': [free_apps['Rating'].mean(), paid_apps['Rating'].mean()],
    'Reviews': [free_apps['Reviews'].mean(), paid_apps['Reviews'].mean()],
    'Installs': [free_apps['Installs'].mean(), paid_apps['Installs'].mean()]
}, index=['Free', 'Paid'])
```

```python
average_values.plot(kind='bar', figsize=(10, 6), color=['lightgreen', 'lightcoral'])
plt.title('Comparison of Free vs. Paid Apps')
plt.xlabel('App Type')
plt.ylabel('Average Value')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
df_ver3['Size'] = pd.to_numeric(df_ver3['Size'], errors='coerce')
df_ver3['Price'] = df_ver3['Price'].replace('[\$,]', '', regex=True).astype(float)
df_ver3['Rating'] = pd.to_numeric(df_ver3['Rating'], errors='coerce')
```

```python
high_rated_apps = df_ver3[df_ver3['Rating'] >= 4.0]
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Size Distribution (Histogram)
axes[0].hist(high_rated_apps['Size'].dropna(), bins=30, color='skyblue', edgecolor='black')
axes[0].set_title('Size Distribution of High-Rated Apps')
axes[0].set_xlabel('Size (MB)')
axes[0].set_ylabel('Number of Apps')


# Price Distribution (Histogram)
axes[1].hist(high_rated_apps['Price'].dropna(), bins=30, color='lightgreen', edgecolor='black')
axes[1].set_title('Price Distribution of High-Rated Apps')
axes[1].set_xlabel('Price (USD)')
axes[1].set_ylabel('Number of Apps')

# Type Distribution (Bar chart)
high_rated_apps['Type'].value_counts().plot(kind='bar', color='lightcoral', ax=axes[2])
axes[2].set_title('Type Distribution of High-Rated Apps')
axes[2].set_xlabel('App Type')
axes[2].set_ylabel('Number of Apps')

plt.tight_layout()
plt.show()
```

```python
# Convert 'Last Updated' column to datetime
df_ver3['Last Updated'] = pd.to_datetime(df_ver3['Last Updated'], errors='coerce')
```

```python
# Extract year and month for time series analysis
df_ver3['Year-Month'] = df_ver3['Last Updated'].dt.to_period('M')
```

```python
plt.figure(figsize=(12, 6))
df_ver3['Year-Month'].value_counts().sort_index().plot(kind='line', color='blue')
plt.title('Trends in App Updates Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Number of Updates')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot: Distribution of Android Version Requirements
df_ver3['Android Ver'] = df_ver3['Android Ver'].str.extract('(\d+\.\d+)')
df_ver3['Android Ver'] = pd.to_numeric(df_ver3['Android Ver'], errors='coerce')
```

```python
plt.figure(figsize=(12, 6))
df_ver3['Android Ver'].dropna().value_counts().sort_index().plot(kind='line', color='green')
plt.title('Trends in Android Version Requirements')
plt.xlabel('Android Version')
plt.ylabel('Number of Apps')
plt.tight_layout()
plt.show()
```
